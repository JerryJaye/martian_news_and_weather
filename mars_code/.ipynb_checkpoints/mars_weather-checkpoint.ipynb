```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
url = 'https://static.bc-edx.com/data/web/mars_facts/temperature.html'
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser') 
```

### Extract the Header Row

```python
# Extract all rows of data
# Find the row with the column headers
header_row = soup.find('tr')

# Extract and print each column header
column_headers = [th.get_text() for th in header_row.find_all('th')]

print(column_headers)
```

### Step 3: Store the Data

### Extract the data rows.
Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Hereâ€™s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# extract all data rows
# Create an empty list
data_rows = []

# Find all table rows with class"class 'data-row'
table_rows = soup.find_all('tr', class_='data-row')

# Loop through each row

for row in table_rows:
    # find all 'td' elements in this row
    tds = row.find_all('td')
    
    # Extract text from each 'td' and append to the row list
    row_data = [td.text for td in tds]
    
    # Append this row's data to the main list
    data_rows.append(row_data)
    
```

### Create and Pandas DataFrame

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
df = pd.DataFrame(data_rows, columns=column_headers)
```

```python
# Confirm DataFrame was created successfully
df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


### Examing the data type for each column. Convert data types as instructed.

```python
# Examine data type of each column
df.info()
```

```python
# Change data types for data analysis
df['terrestrial_date'] = pd.to_datetime(df['terrestrial_date'])
df['sol'] = df['sol'].astype('int64')
df['ls'] = df['ls'].astype('int64')
df['month'] = df['month'].astype('int64')
df['min_temp'] = df['min_temp'].astype('float64')
df['pressure'] = df['pressure'].astype('float64')
```

```python
# Confirm type changes were successful by examining data types again
df.info()
```

### What are the average Minimum Temperatures on Mars each month

```python
# 3. What is the average low temperature by month?
average_low_temp_by_month = df.groupby('month')['min_temp'].mean()
print(average_low_temp_by_month)
```

### Plot of Average Minumum Temperature by Month

```python
# Group by 'Month' and calculate the mean of 'min_temp'

average_min_temp_by_month = df.groupby('month')['min_temp'].mean().reset_index()
sorted_avg_min_temp_df = average_min_temp_by_month.sort_values('month').reset_index(drop=True)

# Create a vertical bar chart
plt.figure(figsize=(8, 5))
plt.bar(sorted_avg_min_temp_df['month'], sorted_avg_min_temp_df['min_temp'], color='blue', width=0.6)

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Temperature in Celsius')
plt.title('Average Minimum Temperature by Month')
plt.xticks(range(1, 13))          

# Display the plot
plt.show()
```

### Plot of coldest to hottest months at Curiosity's location.

```python
# Identify the coldest and hottest months in Curiosity's location
import numpy as np

# Step 1: Aggregate data by month and calculate mean min_temp
monthly_avg_temp = df.groupby('month')['min_temp'].mean().reset_index()

# Step 2: Sort the DataFrame by mean min_temp in ascending order (for coldest to warmest)
monthly_avg_temp_sorted = monthly_avg_temp.sort_values(by='min_temp')

# Step 3: Plot the sorted data
plt.figure(figsize=(8, 5))
plt.bar(monthly_avg_temp_sorted['month'].astype(str), monthly_avg_temp_sorted['min_temp'], color='blue', width=0.6)
plt.title('Average Minimum Temperature by Martian Month (Sorted)')
plt.xlabel('Martian Month')
plt.ylabel('Average Minimum Temperature (Â°C)')
plt.xticks  # Rotate month labels for better readability
plt.show()
```

### List and Plot of average Pressure by month

```python
# 3. What is the average pressure by month?
average_pressure_by_month = df.groupby('month')['pressure'].mean()
print(average_pressure_by_month)
```

```python
# 4. Average pressure by Martian month
average_pressure_by_month_df = average_pressure_by_month.reset_index()
sorted_pressure_df = average_pressure_by_month_df.sort_values(by='month')

# Define the positions for the bars
num_bars = len(sorted_pressure_df)
bar_positions = np.arange(num_bars)

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(bar_positions, sorted_pressure_df['pressure'], color='blue', width=0.6)

# Setting tick positions and labels
plt.xticks(bar_positions, labels=sorted_pressure_df['month'])

# Adding labels and title
plt.xlabel('Martian Month')
plt.ylabel('Average Pressure (Pa)')
plt.title('Average Pressure by Martian Month')

plt.show()
```

### Visual Representatioln of Martian temperature cycles against Earth days.

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
# Calculate the number of terrestrial days since the start
df['days_since_start'] = (df['terrestrial_date'] - df['terrestrial_date'].iloc[0]).dt.days

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(df['days_since_start'], df['min_temp'], label='Minimum Temperature', color='blue')

# Setting the range for the Y-axis as per your specification
plt.ylim(-92, -60)

# Adding labels and title
plt.title('Minimum Temperatures on Mars Over Time')
plt.xlabel('Terrestrial Days from Start')
plt.ylabel('Minimum Temperature (Â°C)')
plt.legend()
plt.grid(color='lightgray', linestyle='--', linewidth=0.5)

# Show plot
plt.show()
```
